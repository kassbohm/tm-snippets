```python
from sympy import *

# S 87

pprint("\nAufgabe 1")
s, t = var("s, t")

pprint("\na)")
l1, t1, r1, s1 =  2,  3, 11, -3
l2, t2, r2, s2 =  6, -2,  8,  0
l3, t3, r3, s3 = -5,  1,  8, -4

eq1 = Eq(  l1 + t1*t, r1 +s1*s )
eq2 = Eq(  l2 + t2*t, r2 +s2*s )
eq3 = Eq(  l3 + t3*t, r3 +s3*s )

sol = solve([eq1, eq2])
sol_s, sol_t = sol[s], sol[t]
eq3 = eq3.subs([ (s, sol_s), (t, sol_t) ])

pprint(sol)
pprint(eq3)

pprint("\nb)")
l1, t1, r1, s1 =  3,  3, 10, -4
l2, t2, r2, s2 = 17, -6, -9,  2
l3, t3, r3, s3 =  1,  9, -5,  1

eq1 = Eq(  l1 + t1*t, r1 +s1*s )
eq2 = Eq(  l2 + t2*t, r2 +s2*s )
eq3 = Eq(  l3 + t3*t, r3 +s3*s )

sol = solve([eq1, eq2])
sol_s, sol_t = sol[s], sol[t]
eq3 = eq3.subs([ (s, sol_s), (t, sol_t) ])

pprint(sol)
pprint(eq3)

# a)
# {s: 4, t: -1}
# False
#
# b)
# {s: -2, t: 5}
# False

# ---

pprint("\nAufgabe 2")

pprint("\na)")
s, t = var("s, t")
p1 = Matrix([1,7,9])
d1 = Matrix([0,1,2])
p2 = Matrix([3,5,5])
d2 = Matrix([1,1,2])
x1 = p1 + s*d1
x2 = p2 + t*d2
eq = Eq(x1, x2)

sol = solve(Eq(x1,x2),[s,t])
pprint(sol)
sol_s, sol_t = sol[s], sol[t]
x1 = x1.subs([(s, sol_s),])
pprint(x1)

pprint("\nb)")
r, s = var("r, s")
p1 = Matrix([9,0,6])
d1 = Matrix([3,2,1])
p2 = Matrix([7,-2,2])
d2 = Matrix([1,1,2])
x1 = p1 + r*d1
x2 = p2 + s*d2
eq = Eq(x1, x2)

sol = solve(Eq(x1,x2),[r,s])
pprint(sol)
sol_r, sol_s = sol[r], sol[s]
x1 = x1.subs([(r, sol_r),])
pprint(x1)

# a)
# {s: -4, t: -2}
# ⎡1⎤
# ⎢ ⎥
# ⎢3⎥
# ⎢ ⎥
# ⎣1⎦
#
# b)
# {r: 0, s: 2}
# ⎡9⎤
# ⎢ ⎥
# ⎢0⎥
# ⎢ ⎥
# ⎣6⎦


# ---

pprint("\nAufgabe 3")
pprint("\na)")
t, s = var("t, s")
p1 = Matrix([5,0,1])
d1 = Matrix([2,1,-1])
p2 = Matrix([-6,-22,-22])
d2 = Matrix([5,8,7])
x1 = p1 + t*d1
x2 = p2 + s*d2
eq = Eq(x1, x2)

sol = solve(Eq(x1,x2),[t,s])
pprint(sol)
sol_t, sol_s = sol[t], sol[s]
x1 = x1.subs([(t, sol_t),])
pprint(x1)

pprint("\nb)")
r, t = var("r, t")
p1 = Matrix([ 1, 2, 1])
d1 = Matrix([ 2, 0, 1])
p2 = Matrix([ 2, 3, 4])
d2 = Matrix([ 0, 1,-1])
x1 = p1 + r*d1
x2 = p2 + t*d2
eq = Eq(x1, x2)

sol = solve(Eq(x1,x2),[r,t])
pprint(sol)
  
# a)
# {s: 3, t: 2}
# ⎡9 ⎤
# ⎢  ⎥
# ⎢2 ⎥
# ⎢  ⎥
# ⎣-1⎦
#
# b)
# []
```
