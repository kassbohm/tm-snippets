```python
from sympy.physics.units import *
from sympy import *

E, A, l, F = var("E, A, l, F")

# sub_list = [
#     (E , 10000 *mega*pascal),
#     (A , 25 *mm*mm),
#     (l , 250 *mm),
#     (F , 1000 *newton),
#     ]
#
c = E*A/l

F = var("F")
c1,c2,c3,c4,c5 = var("c1,c2,c3,c4,c5")
u1,u2,u3,u4 = var("u1,u2,u3,u4")
F1,F2,F3,F4 = var("F1,F2,F3,F4")

u1 = 0
F2 = 0
F3 = 0
F4 = F

sub_list = [
    (c1, c),
    (c2, c),
    (c3, c),
    (c4, 2*c),
    (c5, c),
    ]

u = Matrix([u1,u2,u3,u4])
f = Matrix([F1,F2,F3,F4])
K = Matrix([
    [ c1, -c1, 0, 0 ],
    [ -c1, c1+c2+c3+c4, -c2-c3, -c4 ],
    [0,-c2-c3,c2+c3+c5,-c5],
    [0,-c4,-c5,c5+c4]
    ])

K = K.subs(sub_list)

pprint("\nSolution:")
sol = solve(Eq(K*u,f),[u2,u3,u4,F1], dict=True)
sol = sol[0]
for x in sol:
    pprint("\n")
    tmp = str(x)+":"
    pprint(tmp)
    tmp = sol[x]
    pprint(tmp)

# Solution:
#
# u2:
# F⋅l
# ───
# A⋅E
#
# u3:
# 9⋅F⋅l
# ─────
# 8⋅A⋅E
#
# u4:
# 11⋅F⋅l
# ──────
# 8⋅A⋅E
#
# F1:
# -F
```
