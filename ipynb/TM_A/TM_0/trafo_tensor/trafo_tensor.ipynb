```python
# https://colab.research.google.com/github/kassbohm/tm-snippets/blob/master/ipynb/TM_A/TM_0/trafo_tensor/trafo_tensor.ipynb

# On input:
#    (x,y)-components of the symm. tensor.
#    Angular position phi.

# On output:
#    (\bar x, \bar y)`-components.
#    Principal values and resp. angles.

# User input starts here.

# Tensor components in their resp. unit:
(Txx, Txy, Tyy)  = (-1, 4, 5)
# Angle in deg:
phi = 30

# User input ends here.

from sympy import *
import decimal
from decimal import Decimal as DX

def iso_round(obj, pv, rounding=decimal.ROUND_HALF_EVEN):
    import sympy
    """
    Rounding acc. to DIN EN ISO 80000-1:2013-08
    place value = Rundestellenwert
    """
    assert pv in set([
        # place value   #  round to:
        1,              #  1
        0.1,            #  1st digit after decimal
        0.01,           #  2nd
        0.001,          #  3rd
        ])
    try:
        tmp = DX(str(float(obj)))
        obj = tmp.quantize(DX(str(pv)), rounding=rounding)
    except:
        for i in range(len(obj)):
            tmp = DX(str(float(obj[i])))
            obj[i] = tmp.quantize(DX(str(pv)), rounding=rounding)
    return obj

prec = 0.01
f = lambda x: iso_round(x,prec)

print(u'\nComponents wrt (x, y):')
T = Matrix([ [Txx, Txy], [Txy, Tyy] ])
pprint(T)

print(u"\nComponents wrt (x\u0304, y\u0304)", end="")
print(" for φ = ", end="")
print(iso_round(phi,prec), end="")
print(" deg:")

phi *= pi/180
c, s = cos(phi), sin(phi)
R = Matrix([ [c, s] , [-s, c] ])
Rt = R.transpose()
tmp = R*T*Rt
tmp = tmp.applyfunc(f)
pprint(tmp)

pprint("\n(λ₁, λ₂):")
ev = T.eigenvals()
tmp = Matrix([max(ev), min(ev)])
tmp = tmp.applyfunc(f)
pprint(tmp)

pprint("\n(φ₁, φ₂) / deg:")
t1 = Txy
t2 = (Txx - min(ev))
if t1==0:
    p1 = pi/2
else:
    p1 = atan(t1/t2)
p1 *= 180/pi
p2 = p1 + 90
p = Matrix([iso_round(p1,prec), iso_round(p2,prec)])
pprint(p)

# Components wrt (x, y):
# ⎡-1  4⎤
# ⎢     ⎥
# ⎣4   5⎦
#
# Components wrt (x̄, ȳ) for φ = 30.00 deg:
# ⎡3.96  4.6 ⎤
# ⎢          ⎥
# ⎣4.6   0.04⎦
#
# (λ₁, λ₂):
# ⎡7.0 ⎤
# ⎢    ⎥
# ⎣-3.0⎦
#
# (φ₁, φ₂) / deg:
# ⎡63.43 ⎤
# ⎢      ⎥
# ⎣153.43⎦
```
