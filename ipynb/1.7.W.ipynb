```python
from sympy.physics.units import *
from sympy import *

# Given symbols:
q, b, beta = var("q, b, beta")

# Given in terms of these symbols:
h = beta * b

x = var("x")
z = 4 * beta/b *x**2

# --- a ---
zp = diff(z,x)
pprint("\n\nz':")
pprint(zp)

# --- b ---
pprint("\n\nSupport reactions Ah, Av, H:")
Ah, Av, H = var("Ah, Av, H")

R = q*b/2
eq1 = Eq(H, Ah)
eq2 = Eq(Av, R)
eq3 = Eq(h*H, b/4*R)

sol=solve([eq1, eq2, eq3], [Ah, Av, H])
pprint(sol)
H = sol[H]

# --- c ---
pprint("\n\nSection Forces/Moment N, Q, M:")
phi = atan(zp)
Rx = q*x
c, s = cos(phi), sin(phi)

N, Q, M = var("N, Q, M")

eq1 = Eq(N*c - Q*s - H)
eq2 = Eq(Q*c + N*s - Rx)
eq3 = Eq(Rx*x/2 - H*z - M)

sol = solve([eq1, eq2, eq3], [N, Q, M])
pprint(sol)
N = sol[N]

# --- d ---
pprint("\n\nN / H:")
pprint(N / H)

pprint("\n\nN / R:")
pprint(N / R)

# z':
# 8⋅β⋅x
# ─────
#   b
#
#
# Support reactions Ah, Av, H:
# ⎧    b⋅q      b⋅q     b⋅q⎫
# ⎨Ah: ───, Av: ───, H: ───⎬
# ⎩    8⋅β       2      8⋅β⎭
#
#
# Section Forces/Moment N, Q, M:
# ⎧                   ______________      ⎫
# ⎪                  ╱         2  2       ⎪
# ⎪                 ╱      64⋅β ⋅x        ⎪
# ⎪         b⋅q⋅   ╱   1 + ────────       ⎪
# ⎨               ╱            2          ⎬
# ⎪             ╲╱            b           ⎪
# ⎪M: 0, N: ────────────────────────, Q: 0⎪
# ⎪                   8⋅β                 ⎪
# ⎩                                       ⎭
#
#
# N / H:
#       ______________
#      ╱         2  2
#     ╱      64⋅β ⋅x
#    ╱   1 + ────────
#   ╱            2
# ╲╱            b
#
#
# N / R:
#       ______________
#      ╱         2  2
#     ╱      64⋅β ⋅x
#    ╱   1 + ────────
#   ╱            2
# ╲╱            b
# ────────────────────
#         4⋅β
```
