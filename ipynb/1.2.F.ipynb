```python
from sympy.physics.units import *
from sympy import *


newton = kg*m/s**2

# Given symbols:
F1, F2 = var("F1, F2", real=true)
a1, a2 = var("a1, a2", real=true)

sub_list=[
    ( F1, 6 *newton  ),
    ( F2, 2 *newton  ),
    ( a1,  60*pi/180 ),
    ( a2,  45*pi/180 ),
    ]

p1 = pi + a1
p2 = -a2

c1, s1 = cos(p1), sin(p1)
c2, s2 = cos(p2), sin(p2)

e1 = Matrix([c1, s1])
e2 = Matrix([c2, s2])

F1 = F1 * e1
F2 = F2 * e2

R = F1 + F2

pprint("\n(x,y)-comps of R / N:")
tmp = R
tmp /= newton
tmp = tmp.subs(sub_list)
prec = 2
pprint(N(tmp,prec))

pprint("\nMagnitude of R / N:")
Rmag = R.norm()
tmp = Rmag
tmp = tmp.subs(sub_list)
tmp /= newton
pprint(N(tmp, prec))

pprint("\nphi_R / deg:")
prec = 3
Rx, Ry = R[0], R[1]
Rx = Rx.subs(sub_list)
Ry = Ry.subs(sub_list)
Rx = Rx.simplify()
Ry = Ry.simplify()

tmp = atan2(Ry, Rx)
tmp = deg(tmp)
pprint(N(tmp, prec))

# tmp = acos(-Ry/Rmag)
# tmp = tmp.subs(sub_list)
# tmp = deg(tmp)
# tmp += 90
# tmp *= -1
# pprint(N(tmp, prec))
#
# tmp = asin(-Rx/Rmag)
# tmp = tmp.subs(sub_list)
# tmp = deg(tmp)
# tmp += 90
# tmp *= -1
# pprint(N(tmp, prec))
#
# tmp = atan(-Rx/-Ry)
# tmp = tmp.subs(sub_list)
# tmp = deg(tmp)
# tmp += 90
# tmp *= -1
# pprint(N(tmp, prec))
```
